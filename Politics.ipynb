IMPORT LIBRARIES


```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import plotly.graph_objects as go
import plotly.express as  xp
```

LOADING DATA



```python
modi =pd.read_csv("/content/modi_reviews.csv")
rahul =pd.read_csv("/content/rahul_reviews.csv")
```

```python
TextBlob(rahul['Tweet'][2]).sentiment
```

```python
modi['Tweet']=modi['Tweet'].astype(str)
rahul['Tweet']=rahul['Tweet'].astype(str)

def findpolarity(review):
  return TextBlob(review).sentiment.polarity
```

```python
modi['Polarity']=modi['Tweet'].apply(findpolarity)
rahul['Polarity']=rahul['Tweet'].apply(findpolarity)
```

```python
modi.head()
```

```python
rahul.head()
```

```python
modi['Label']=np.where(modi['Polarity']>0,'Positive','Negative')
modi['Label'][modi['Polarity']==0]='Neutral'

rahul['Label']=np.where(rahul['Polarity']>0,'Positive','Negative')
rahul['Label'][rahul['Polarity']==0]='Neutral'
```

```python
modi
```

```python
neutral_modi=modi[modi['Polarity']==0.000]
remove_neutral_modi=modi['Polarity'].isin(neutral_modi['Polarity'])
modi.drop(modi[remove_neutral_modi].index,inplace=True)
```

```python
modi.shape
```

```python
rahul.shape
```

```python
# making row count to 9000
# for modi
np.random.seed(10)
remove_n=8482
drop_indices=np.random.choice(modi.index,remove_n,replace=False)
df_modi=modi.drop(drop_indices)
```

```python
df_modi.shape
```

```python
# making row count to 9000
# for rahul
np.random.seed(10)
remove_n=8482
drop_indices=np.random.choice(rahul.index,remove_n,replace=False)
df_rahul=rahul.drop(drop_indices)
```

```python
df_rahul.shape
```

prediction for election



```python
modi_count = df_modi.groupby('Label').count()
modi_count
```

```python
nef_modi = (modi_count['Polarity'].loc['Negative'] / 1000) * 100
pos_modi = (modi_count['Polarity'].loc['Positive'] / 1000) * 100
```

```python
rahul_count =  df_rahul.groupby('Label').count()
nef_rahul = (rahul_count['Polarity'].loc['Negative'] / 1000) * 100
pos_rahul = (rahul_count['Polarity'].loc['Positive'] / 1000) * 100
```

```python
rahul_count =  df_rahul.groupby('Label').count()
rahul_count
```

```python
politicians = ['Modi', 'Rahul']
neg_list = [nef_modi, nef_rahul]
pos_list = [pos_modi, pos_rahul]

fig = go.Figure(data=[
    go.Bar(name='Negative', x=politicians, y=neg_list),
    go.Bar(name='Positive', x=politicians, y=pos_list)
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()
```
